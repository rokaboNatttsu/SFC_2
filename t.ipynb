```julia
using StatsPlots
```

```julia
T = 100
λ_e = 0.5
X = 100.0*exp.([i for i =1/T:1/T:1]) .* (1.0 .+ 0.01*randn(T))
Xe = fill(100.0, T)
for t = 1:T
    tm1, tm2 = (T + t - 1 - 1)%100 + 1, (T + t - 2 - 1)%100 + 1
    Xe[t] = (λ_e*X[tm1] + (1 - λ_e)*Xe[tm1])*X[tm1]/X[tm2]
end
```

```julia
plot(X[end-10:end], label="X")
plot!(Xe[end-10:end], label="Xe")
```

```julia
plot(X[end-10:end] - Xe[end-10:end], label="Δ")
```
